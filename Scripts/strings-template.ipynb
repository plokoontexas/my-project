### Store and print a variable

```python
# define some strings
first = 'Lucy'
last = 'Dacus'

# concatenate those strings
full = first + last 

# print the strings

print(full + " is the best")
```

### Get the length of a string

```python
# the long (but readable) way

name_length = len(full)

print(name_length)

# a more concise way to do it, functions within functions
print(len(full))
```

### Print strings and numbers

```python
# print strings and integers

print("There are " + str(name_length) + " characters in Lucy's name")

# print strings

print("There are", str(name_length), "characters in Lucy's name")

# formatted print

print(f"There are {name_length} characters in Lucy's name")
```

### Problem #2: Ask the user for a big number. If the user enters 1000000 (one million), the output will be:

### Problem #4: Tip Calculator

Ask the user for the bill amount (float) and tip percentage (integer). Calculate the tip and total, then print them neatly.

Enter the bill amount: `42.75`

Enter the tip percentage: `18`

Bill: `$42.75`

Tip (18%): `$7.70`

\--------------------

Total: `$50.45`


```python
# problem 5 solution

Total = int(input("Enter bill amount: "))
Percent = int(input("Tip Percentage: "))
Percentage = (Percent / 100)

Tip = (Total * Percentage)

Bill = (Total + Tip)

print("Bill Amount:", Total,"Tip Percentage:", Percentage,"Tip Amount:", Tip,"Final Bill:", Bill)

print(f"Bill Amount: ${Total} Tip Percentage: {Percentage}% Tip Amount: ${Tip} Final Bill: ${Bill}")





```
